```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonar_data=pd.read_csv(r"C:\Users\jerem\Desktop\ml data\sonar data.csv",header=None)
sonar_data
```

#getting the number of rock records and the number of mines in our data

```python
sonar_data[60].value_counts()
```

```python
#seperating the data from the labels
```

```python
x=sonar_data.drop(columns=60,axis=1)
print(x)
y=sonar_data[60]
print(y)
```

```python
#spliting the data into training set and test set
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.1,stratify=y,random_state=2)
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
#training our model
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
#evaluating our model
```

```python
train_prediction=model.predict(x_train)
train_accuracy=accuracy_score(train_prediction,y_train)
print(train_accuracy)
test_prediction=model.predict(x_test)
test_accuracy=accuracy_score(test_prediction,y_test)
print(test_accuracy)
```

```python
#creating  a rock vs mine predicting system
```

```python
input_data = input('Input Data :')

# Split the input string by commas, filter out empty strings, and convert each part to a float
input_data_list = [float(i) for i in input_data.split(',') if i.strip()]
array_data = np.asarray(input_data_list)
reshaped_data = array_data.reshape(1, -1)

prediction = model.predict(reshaped_data)
print(f'Prediction: {prediction[0]}')
```
